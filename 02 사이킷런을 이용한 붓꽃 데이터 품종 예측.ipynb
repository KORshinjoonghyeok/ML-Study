```python
# 붓꽃 데이터 로드 function
from sklearn.datasets import load_iris
# 분류 알고리즘 
from sklearn.tree import DecisionTreeClassifier
# 학습데이터, 테스트데이터 분리하는 함수 
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import pandas as pd
```

```python
# 붓꽃 데이터 세트를 로딩
iris = load_iris()
type(iris)
```

```python
iris
```

```python
# iris이 데이터 추출
iris_data = iris.data
```

```python
iris_data
```

```python
# iris.target key : 레이블(답) 데이터 추출
iris_label = iris.target
iris_label
```

```python
# 품종명 추출
iris.target_names
```

```python
# 추출한 데이터 -> 데이터프레임 생성
iris_df = pd.DataFrame(
      data=iris_data
)
iris_df.head(3)
```

```python
# 추출한 데이터 -> 데이터프레임 생성
iris_df = pd.DataFrame(
      data=iris_data
    , columns=iris.feature_names # 컬럼명
)
iris_df.head(3)
```

```python
# 레이블(답) 추가
iris_df['label'] = iris.target
iris_df.head(3)
```

```python
# 전체 데이터 : 150개
# 학습 데이터 : 120개, 테스트 데이터 : 30개(0.2) 분리
# 학습 레이블 : 120개, 테스트 레이블 : 30개 분리
# 학습데이터 , 테스트데이터, 학습답, 테스트답
X_train    , X_test,       y_train, y_test= train_test_split(
      iris_data # 분리할 데이터
    , iris_label # 분리할 답
    , test_size=0.2 # 테스트데이터의 크기 지정
    , random_state=11 # random_seed 동일
)
```

```python
# 학습시 사용될 학습데이터, 학습 답
X_train, y_train
```

```python
# 테스트시 사용될 데스트데이터, 테스트 답
X_test, y_test
```

```python
# 모델 : 학습데이터와 학습답으로 학습한 결과
# 분류로 학습 진행
# DecisionTreeClassifier : 분류 처리 객체만 생성 -> 학습X
dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf
```

```python
# 학습데이터(X_train), 답(y_train)을 넣어서 학습 수행
dt_clf.fit(X_train, y_train)
# dt_clf라는 객체안에 모델(학습결과 : 식) 생성
```

```python
# 테스트데이터(X_test)를 모델 넣어서 예측 결과 반환
pred = dt_clf.predict(X_test)
pred
```

```python
# 0번째 테스트데이터, 예측값,     원래값
X_test[0],            pred[0],    y_test[0]
```

```python
pred == y_test
```

```python
np.sum(pred == y_test)
```

```python
# 정확도 계산

28/30 * 100
```

```python
# accuacy_score : 정확도 계산 함수
# y_test(원래값) == pred(예측결과값)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
print('예측 정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))
```
